```python
import csv
import numpy as np
import math as m
import datetime

#f = open('BM869_CTS_TC_19april.csv')
f = open('BM869_CTS_TC.csv')

csvreader = csv.reader(f)
header = []
header = next(csvreader)
rows = []
nr_samples=0
for row in csvreader:
  rows.append(row)
  nr_samples = nr_samples + 1
T    = np.empty(nr_samples,dtype=np.float64)
Vref = np.empty(nr_samples,dtype=np.float64)
Tset = np.empty(nr_samples,dtype=float)
Tact = np.empty(nr_samples,dtype=float)
for i in range (nr_samples):
    T[i]    = (float(rows[i][0])- float(rows[0][0]))/60
    Vref[i] = float(rows[i][1])
    Tset[i] = float(rows[i][2])
    Tact[i] = float(rows[i][3])
    if Tact[i] <= 0:
        Tact[i] = Tact[i-1]
```

```python


dt = datetime.datetime.fromtimestamp(T[0])
print (dt)
```

```python
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots(figsize=(15, 15))

color = 'tab:blue'
ax1.set_xlabel('time (s)')
ax1.set_ylabel('Vref', color=color)
ax1.plot(T, Vref, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:red'
ax2.set_ylabel('Tact', color=color)  # we already handled the x-label with ax1
ax2.plot(T, Tact, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
